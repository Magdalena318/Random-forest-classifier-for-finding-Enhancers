```python
from itertools import product
import math
from Bio.Seq import reverse_complement
from sklearn.ensemble import RandomForestClassifier
import numpy
import sys
```

```python
features = []
prev_id = 0
nucs = ["A", "T", "C", "G"]
ids = {}
```

```python
X = []
Y = []
```

```python
step = 750
frame = 1500
```

```python
def count_frequency(seq):
    freq=numpy.zeros(len(features))
    for x in range(len(seq)-3):
        prev = seq[x:x+4]
        freq[ids[prev]]+=1
    
    freq/= len(seq)-3
    return(freq)
```

```python
def combinations():
  global prev_id, features, ids
  for c in product(nucs, repeat=4):
    c = ''.join(c)
    if (reverse_complement(c) in ids):
        ids[c]= ids[reverse_complement(c)]
    else : 
        features.append(c)
        ids[c] = prev_id 
        prev_id += 1
```

```python
def get_train_data(file):
  with open(file, 'r') as f:
    for count, line in enumerate(f, start=1):
        if count % 2 == 0:
            X.append(count_frequency(line.rstrip().upper()))
            if(file == 'vista1500.txt'):
              Y.append(1)
            elif(file == 'randoms1500.txt'):
              Y.append(0)
```

```python
def get_test_data(file):
  f= open(file,"r")
  f.readline()
  data = f.read().replace('\n', '').upper()
  f.close()

  test = []

  for x in range(math.floor((len(data)-frame)/step)):
      start = x*step
      tmp = data[start:start+frame]
      if ("N" not in tmp):
          test.append(count_frequency(tmp))

  return test
```

```python
def main():
  combinations()
  get_train_data('randoms1500.txt')
  get_train_data('vista1500.txt')

  #train
  rfc = RandomForestClassifier()
  rfc.fit(X, Y)

  
  #test
  test = get_test_data("chr21.fa")
  predict = rfc.predict(test)
  numpy.set_printoptions(threshold=sys.maxsize)
  print(predict)

  

main()
```
